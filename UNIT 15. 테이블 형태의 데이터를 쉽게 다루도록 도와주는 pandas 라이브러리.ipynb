## 행 방향 축을 기준으로 한 연산
* A열의 값을 B열의 값으로 나눈 후, 그 결과를 새로 만든 D 열에 저장

```python
import pandas as pd
import numpy as np

index = pd.date_range('1/1/2000', periods=8)
df = pd.DataFrame(np.random.rand(8,3), index = index, columns=list('ABC'))
df['D']= df['A']/df['B'] # A열의 값을 B열의 값으로 나눈 후, 그 결과를 새로 만든 D 열에 저장
df
```

## 열 방향 축을 기준으로 한 연산
* df 데이터 프레임에서 행 우선으로 합을 구하고, 그 결과를 E라는 열을 생성해서 저장하는 코드

```python
import pandas as pd
import numpy as np

index = pd.date_range('1/1/2000', periods=8)
df = pd.DataFrame(np.random.rand(8,3), index = index, columns=list('ABC'))
df['D'] = df['A']/df['B']
df['E']=np.sum(df, axis=1) # 행 우선 계산 값을 E열에 저장
df.head()
```

## 전체 데이터에 대해서 열 우선 계산을 하는 방법
* 전체 데이터를 A열의 값으로 뺄 때는 sub() 함수를 사용

```python
import pandas as pd
import numpy as np

index = pd.date_range('1/1/2000', periods=8)
df = pd.DataFrame(np.random.rand(8,3), index = index, columns=list('ABC'))
df['D'] = df['A']/df['B']
df['E']=np.sum(df, axis=1) 
df = df.sub(df['A'],axis=0) # A열의 데이터를 기준으로 열 우선 계산
df.head() # DataFrame의 첫 5개 행을 출력
```

* 데이터 전체를 C열의 값으로 나눌 때는 div() 함수를 사용
* 그리고 데이터 프레임을 csv파일로도 쉽게 저장 할 수 있다. 

```python
import pandas as pd
import numpy as np

index = pd.date_range('1/1/2000', periods=8)
df = pd.DataFrame(np.random.rand(8,3), index = index, columns=list('ABC'))
df['D'] = df['A']/df['B']
df['E']=np.sum(df, axis=1) 
df = df.sub(df['A'],axis=0)
df = df.div(df['C'],axis=0) # C열 데이터를 기준으로 열 우선 계산 
df.to_csv('test.csv') # 데이터 프레임을 test.csv 파일로 저장
df.head() 
```

## 4-5| 궁금한 지역의 인구 구조와 가장 비슷한 인구 구조를 가진 지역 시각화하기

> 먼저, df에서 df2를 제외한 전체 DataFrame에서 지정한 지역(df2)과 인구 구조가 가장 유사한 상위 5개 지역을 찾기 

* df2.iloc[0]: df2에서 첫 번째 행을 추출합니다.
* df.sub(df2.iloc[0], axis=1): df2의 첫 번째 행을 기준으로 df의 모든 열의 값을 뺍니다. 이때, axis=1을 사용하여 열 우선 계산이 수행됩니다.
* np.power(df.sub(df2.iloc[0], axis=1), 2): 이전 단계에서 계산된 결과에 대해, 모든 원소의 값을 제곱합니다.
* sum(axis=1): 각 행(row)의 모든 열(column)의 값을 더합니다.
* sort_values(): 값이 작은 순서대로 정렬합니다.
* index[:5]: 상위 5개 지역의 인덱스(index)를 추출합니다.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


plt.rc('font', family='Malgun Gothic')
df = pd.read_csv('age.csv',encoding='cp949', index_col=0)
df = df.div(df['2019년02월_계_총인구수'], axis=0)
del df['2019년02월_계_총인구수'], df['2019년02월_계_연령구간인구수']

name = input('원하는 지역의 이름을 입력해주세요 : ')
a = df.index.str.contains(name)
df2=df[a]

df.loc[np.power(df.sub(df2.iloc[0], axis=1), 2).sum(axis=1).sort_values().index[:5]].T.plot()
       
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


plt.rc('font', family='Malgun Gothic')
df = pd.read_csv('age.csv',encoding='cp949', index_col=0)
df = df.rename(columns=lambda x: x.replace('2019년02월_계_', '')) # 2019년02월_계_ 없애기 
df = df.div(df['총인구수'], axis=0)
del df['총인구수'], df['연령구간인구수']

name = input('원하는 지역의 이름을 입력해주세요 : ')
a = df.index.str.contains(name)
df2=df[a]

x = df.sub(df2.iloc[0], axis=1)

y = np.power(x,2)
z = y.sum(axis=1)

i = z.sort_values().index[:5]
df.loc[i].T.plot()

plt.title(name+'동과 비슷한지역')
plt.show()


```
